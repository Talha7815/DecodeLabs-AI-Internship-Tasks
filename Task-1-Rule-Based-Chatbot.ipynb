```python
# Rule-Based AI Chatbot - GOOGLE COLAB VERSION
# Run this cell, then type your messages in the input box below

def chatbot():
    responses = {
        # Greetings
        'hello': 'Hi there! Welcome to DecodeLabs. How can I help you today?',
        'hi': 'Hello! Ready to build some AI?',
        'hey': 'Hey there! What brings you here?',

        # Questions
        'how are you': "I'm functioning perfectly! Just a set of rules doing their job.",
        'what is your name': "I'm Project 1, your rule-based chatbot from DecodeLabs.",
        'who made you': "The interns at DecodeLabs built me as their first AI project!",

        # Help
        'help': "I can respond to greetings, answer basic questions, and follow commands. Try saying 'hello', 'what is ai', or 'bye'!",
        'what can you do': "I respond to specific keywords using a dictionary lookup! O(1) efficiency!",

        # AI Knowledge
        'what is ai': "AI is making machines think like humans. But right now, I'm just following rules - that's the foundation!",
        'what is rule based ai': "Rule-based AI uses explicit if-else logic. No learning, just pure deterministic decision-making.",

        # Fun
        'tell me a joke': "Why did the AI go to therapy? It had too many if-else conflicts!",
        'thank you': "You're welcome! Keep building amazing things.",
        'thanks': "Anytime! Happy coding.",

        # Exit
        'bye': 'Goodbye! Keep building. See you in Project 2!',
        'exit': 'Exiting the chat. Remember: Rules first, then learning!',
        'quit': 'Quitting... Your rule-based foundation is solid now!'
    }

    print("\n" + "="*50)
    print("🤖  DecodeLabs Rule-Based AI Chatbot  🤖")
    print("="*50)
    print("Type 'help' to see what I can do.")
    print("Type 'bye', 'exit', or 'quit' to leave.")
    print("-"*50)

    while True:
        # This works perfectly in Colab
        user_input = input("\nYou: ").strip().lower()

        if not user_input:
            print("Bot: I didn't catch that. Please say something!")
            continue

        if user_input in ['bye', 'exit', 'quit']:
            print(f"Bot: {responses[user_input]}")
            print("\n" + "="*50)
            print("Session ended. Your AI journey continues...")
            print("="*50)
            break

        reply = responses.get(user_input, "I don't understand that yet. Try 'help' to see what I know!")
        print(f"Bot: {reply}")

# Run the chatbot
chatbot()
```
